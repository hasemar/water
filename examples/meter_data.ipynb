```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('meter_data.xlsx',sheetname='Sheet2')
```

```python

new_colNames = ['DEVICE_ID',
                'CURRENT_READ',
                'DATE',
                '00:00',
                '01:00',
                '02:00',
                '03:00',
                '04:00',
                '05:00',
                '06:00',
                '07:00',
                '08:00',
                '09:00',
                '10:00',
                '11:00',
                '12:00',
                '13:00',
                '14:00',
                '15:00',
                '16:00',
                '17:00',
                '18:00',
                '19:00',
                '20:00',
                '21:00',
                '22:00',
                '23:00',
                ]
```

```python
df.info()
```

```python
df.columns = new_colNames
```

```python
df.info()
```

```python
df.head()
```

```python
by_device = df.groupby(['DEVICE_ID']).mean().drop(['CURRENT_READ'], axis=1)
by_device.head()
```

```python
by_time = by_device.pivot_table(columns=['DEVICE_ID'])
```

```python
by_time.head()
```

```python
devices = by_time.columns
fig, ax = plt.subplots(1,1, figsize=(12,8))
x = range(24)

for device in devices:
    ax.scatter(x,y=by_time[device])
    
    

```

```python
n_bins = int(np.ceil(np.sqrt(len(df.index))))
f = plt.figure(figsize=(10,6))
sns.distplot(by_device['06:00'], bins=n_bins, kde=False)
plt.ylabel('n')
plt.xlabel('gals')
```

```python
g = sns.FacetGrid(df, hue="DEVICE_ID", palette='viridis', size=6, aspect=2)
g = g.map(plt.hist, '06:00', bins=n_bins, alpha=0.7)
plt.ylabel('n')
```

```python
heat = plt.figure(figsize=(20,16))
sns.heatmap(by_time.corr(),cmap='coolwarm',annot=False)
```
